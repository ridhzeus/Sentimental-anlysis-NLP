```python
import pandas as pd
import snscrape.modules.twitter as sn
```

```python
em=sn.TwitterUserScraper("narendramodi")
```

```python
li=[]
for i,tweet in enumerate(em.get_items()):
    if i>100:
        break
    li.append([tweet.rawContent,tweet.date,tweet.likeCount,tweet.retweetCount,tweet.replyCount,"Politics"])
```

```python
df=pd.DataFrame(li,columns=['Tweet','Date','Like','Retweet','Comment','Output'])
df.to_csv('politics.csv')
```
